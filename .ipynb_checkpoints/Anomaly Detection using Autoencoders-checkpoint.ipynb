## **Autoencoder**

An autoencoder is a special type of feed-forward neural network that reproduces the input values on the output values and utilizes a hourglass shape to extract the essential information of the input variables. The autoencoder is usually trained using **backpropagation** and **MSE** as loss function. It does not require a target variable $\rightarrow$ **unsupervised leraning**. 

![](AE_single_layer.png)

Consist of an **Encoder**-part which effectively reduces the dimensionality of the input features to a specified level, and an **Decoder**-part which reverses the operations of the encoder. 

The early application of autoencoders is dimensionality reduction. A milestone paper by **[Geoffrey Hinton](https://de.wikipedia.org/wiki/Geoffrey_Hinton)** (2006) showed a trained autoencoder yielding a smaller error compared to the first 30 principal components of a PCA and a better separation of the clusters. Autoencoders also have wide applications in computer vision and image editing. In image coloring, autoencoders are used to convert a black-and-white image to a colored image.

During the process of dimensionality reduction, e.g. using PCA or AE, outliers are identified $\rightarrow$ **outlier detection is a by-product of dimensionlity reduction**.

**Why Autoencoders?** AE perform non-linear transformations with their non-linear activation functions, in constrast to the linear transformations in a PCA $\rightarrow$ allow for greater flexibility

**Aggregate to achieve model stablity** Unsupervised learning techniques are powerful in detecting outliers, but they are also very prone to overfitting and mostly unable to produce stable results. The solution is to train *multiple models* and then aggregate the scores.

### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pyod.models.knn import KNN
from pyod.models.auto_encoder import AutoEncoder
from pyod.utils.data import generate_data
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

### Generate data

```python
contamination = 0.1 # outlier percentage
n_train = 500
n_test = 500
n_features = 25

X_train, X_test, y_train, y_test = generate_data(n_train=n_train, n_test=n_test, n_features=n_features, 
                                                 contamination=contamination, random_state=1234, behaviour="new")

X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
```

### Standardize data

```python
X_train = pd.DataFrame(StandardScaler().fit_transform(X_train))
X_test = pd.DataFrame(StandardScaler().fit_transform(X_test))
```

### Visualize data using PCA

```python
pca = PCA(2)
x_pca = pca.fit_transform(X_train)
x_pca = pd.DataFrame(x_pca)
x_pca.columns=["PC1", "PC2"]

# Plot
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
ax1.scatter(X_train[0], X_train[1], c=y_train, alpha=0.8)
ax1.set_xlabel("x1")
ax1.set_ylabel("x2")
ax1.set_title("Original Data")
ax1.grid()

sp2 = ax2.scatter(x_pca["PC1"], x_pca["PC2"], c=y_train, alpha=0.8)
ax2.set_title("Transformed Data")
ax2.set_xlabel("PC1")
ax2.set_ylabel("PC2")
ax2.grid()
legend1 = ax2.legend(*(sp2.legend_elements()[0], ["Data", "Outlier"]),
                    loc="upper left", title="Datatype")
ax2.add_artist(legend1)
```

### Autoencoder

try different models and take a majority vote since we are in the unsupervised setting

- M1: [25,2,2,25] neurons per layer
- M2: [25,10,2,10,25] neurons per layer
- M3: [25,15,10,2,10,15,25] neurons per layer

## Model 1

```python
clf1 = AutoEncoder(hidden_neurons=[25,2,2,25])
clf1.fit(X_train)
```

### Get outlier score and some visualization

```python
y_train_scores = clf1.decision_scores_   # get the training scores

y_test_scores = pd.Series(clf1.decision_function(X_test))  # get predictions

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,8))
ax1.hist(y_test_scores, bins="auto")
ax1.set_title("Histogram for Model Cfl1 Anomaly Score")
ax1.set_xlabel("x")
ax1.grid()

ax2.scatter(X_test[0], y_test_scores, c=y_test, marker="x")
ax2.set_title("Anomaly Scores")
ax2.grid()
ax2.set_ylabel("x")
legend1 = ax2.legend(*(sp2.legend_elements()[0], ["Data", "Outlier"]),
                    loc="upper left", title="Datatype")
ax2.add_artist(legend1)
```

### Get the summary statistics by cluster

```python
df_test = X_test.copy()
df_test["score"] = y_test_scores
df_test["clusters"] = np.where(df_test["score"]<4, 0, 1)
df_test.groupby('clusters').mean()
```

## Model 2 - Fit and Visualize


```python
clf2 = AutoEncoder(hidden_neurons=[25,10,2,10,25])
clf2.fit(X_train)
```

```python
y_test_scores = pd.Series(clf2.decision_function(X_test))

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,8))
ax1.hist(y_test_scores, bins="auto")
ax1.set_title("Histogram for Model Cfl2 Anomaly Score")
ax1.set_xlabel("x")
ax1.grid()

ax2.scatter(X_test[0], y_test_scores, c=y_test, marker="x")
ax2.set_title("Anomaly Scores")
ax2.grid()
ax2.set_ylabel("x")
legend1 = ax2.legend(*(sp2.legend_elements()[0], ["Data", "Outlier"]),
                    loc="upper left", title="Datatype")
ax2.add_artist(legend1)

# get summary statistics
df_test = X_test.copy()
df_test['score'] = y_test_scores
df_test['cluster'] = np.where(df_test['score']<4, 0, 1)
df_test['cluster'].value_counts()
df_test.groupby('cluster').mean()
```

## Model 3 - Fit and Visualize

```python
# Step 1: Build the model
clf3 = AutoEncoder(hidden_neurons =[25, 15, 10, 2, 10,15, 25])
clf3.fit(X_train)
```

```python
# Predict the anomaly scores
y_test_scores = clf3.decision_function(X_test)  
y_test_scores = pd.Series(y_test_scores)

# Step 2: Determine the cut point
import matplotlib.pyplot as plt
plt.hist(y_test_scores, bins='auto')  
plt.title("Histogram with Model Clf3 Anomaly Scores")
plt.show()

df_test = X_test.copy()
df_test['score'] = y_test_scores
df_test['cluster'] = np.where(df_test['score']<4, 0, 1)
df_test['cluster'].value_counts()

# Step 3: Get the summary statistics by cluster
df_test.groupby('cluster').mean()
```

## Aggregate to Achieve Model Stability

The four main methods are
- Average: average scores of all models
- Maximum of Maximum (MOM)
- Average of Maximum (AOM)
- Maximum of Average (MOA)

When you aggregate the scores form different models, **standardization** is  necessary.

```python
from pyod.models.combination import aom, moa, average, maximization

# Put all the predictions in a data frame
train_scores = pd.DataFrame({'clf1': clf1.decision_scores_,
                             'clf2': clf2.decision_scores_,
                             'clf3': clf3.decision_scores_
                            })

test_scores  = pd.DataFrame({'clf1': clf1.decision_function(X_test),
                             'clf2': clf2.decision_function(X_test),
                             'clf3': clf3.decision_function(X_test) 
                            })
```

```python
train_scores_norm = StandardScaler().fit_transform(X=train_scores)
test_scores_norm = StandardScaler().fit_transform(X=test_scores)
```

```python
# Combination by average
y_by_average = average(test_scores_norm)
plt.hist(y_by_average, bins='auto')  # arguments are passed to np.histogram
plt.title("Combination by average")
plt.show()
```

# Ebner Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pyod.models.knn import KNN
from pyod.models.auto_encoder import AutoEncoder
from pyod.utils.data import generate_data
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from pathlib import Path
```

## Load Data and standardize it

```python
ebner_path = Path('Master-Thesis/Data')
fname = "Table_alpha_Data.txt"
df = pd.read_csv(ebner_path / fname)
X_train = pd.DataFrame(StandardScaler().fit_transform(df[["QDot", "T"]]))
```

## Fit Autoencoder

```python
ae = AutoEncoder(hidden_neurons=[2,1,1,1,2])
ae.fit(X_train)
```

## Plot histogram and anomaly scores

```python
y_train_scores = pd.Series(ae.decision_function(X_train))

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,8))
ax1.hist(y_train_scores, bins="auto")
ax1.set_title("Histogram for Model Cfl2 Anomaly Score")
ax1.set_xlabel("Anomaly Scores")
ax1.grid()

ax2.scatter(X_train[0], y_train_scores, marker="x")
ax2.grid()
ax2.set_ylabel("Anomaly Scores")
ax2.set_xlabel("std. QDot")
```

## Add scores to dataframes

```python
df["score"] = y_train_scores
df["cluster"] = np.where(df["score"]<2.25,0,1)
df["cluster"].value_counts()
```

## Plot the outliers

```python
import plotly.express as px
fig = px.scatter_3d(df, x='QDot', y='T', z='HTC',
              color='cluster')
fig.show()
```
