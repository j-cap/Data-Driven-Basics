```python
%%manim -v WARNING -ql Test

class Test(MovingCameraScene):

    rec_def = Text("Recursive definition by Cox & de Boor", font_size=35)
    f1 = MathTex(r"s(x) = \sum_{j=1}^d B_j^l(x) \beta_j")
    f2 = MathTex(r"B_j^0(x) = \begin{cases} 1 \quad \text{for} \ \kappa_j \le x < \kappa_{j+1} \\ 0 \quad \text{otherwise} \end{cases}")
    f3 = MathTex(r"B_j^l(x) = \frac{x - \kappa_{j-l}}{\kappa_j - \kappa_{j-l}} B_{j-1}^{l-1}(x) + \frac{\kappa_{j+1} - x}{\kappa_{j+1} - \kappa_{j+1-l}} B_j^{l-1}(x)")
    f4 = MathTex(r"t(x_1, x_2) = \sum_{j=1}^{d_1} \sum_{k=1}^{d_2} t_{j,k}^{l_1, l_2}(x_1,x_2) \beta_{j,k}")
    
    color_const = YELLOW
    color_linear = GREEN
    color_cubic = RED
    
    Bspline_headline = Text("B-splines").to_corner(UP+LEFT, buff=1)        
    TP_Bspline_headline = Text("Tensor-product \n B-splines").to_corner(UP+RIGHT, buff=1)
    P_spline_headline = Text("P-splines").to_corner(DOWN+LEFT, buff=1)
    A_spline_headline = Text("A-splines").to_corner(DOWN+RIGHT, buff=1)
       
    
    def construct(self):
        #self.add(*dotted_grid)
        #Bspline_headline = Text("B-splines").move_to([-3,2.5,0])        
        # Create the 4 quadrants
        self.play(Create(self.Bspline_headline), Create(self.TP_Bspline_headline), 
                  Create(self.P_spline_headline), Create(self.A_spline_headline))
        
        # Zoom into the upper left quadrant for B-spline 
        self.camera.frame.save_state()
        
        # move camera to upper left + BSPLINE
        self.play(self.camera.frame.animate.set(width=11, height=11).move_to([-5,3,0]))
        self.plot_Bspline_quadrant()
        self.play(Restore(self.camera.frame))
        
        # move camera to upper right + TENSOR-PRODUCT BSPLINE
        self.play(self.camera.frame.animate.set(width=11, height=11).move_to([13,3,0]))
        
        self.wait(3)
        return
        
    def plot_Bspline_quadrant(self):
        
        self.remove(self.Bspline_headline, self.)
        # define headline
        Bspline_quadrant = Text("B-spline", color=BLUE, font_size=40).move_to([-14,7.6,0], aligned_edge=LEFT)
        GOAL = Text("Goal: 1D Function estimation", font_size=40, color=RED, ).move_to([-3, 7.6, 0], aligned_edge=LEFT)
        # Write headline
        self.play(Write(Bspline_quadrant))
        self.play(Write(GOAL))
        # Write B-spline formula
        self.play(Create(self.f1.next_to(Bspline_quadrant, RIGHT, buff=1)), run_time=2)
        
        self.play(Write(self.rec_def.next_to(Bspline_quadrant, DOWN, aligned_edge=LEFT, buff=1)))
        x = VGroup(self.f2, self.f3).arrange(direction=DOWN, aligned_edge=LEFT).next_to(self.rec_def, RIGHT, buff=-1).scale_in_place(0.7)
        curly_brace_around_rec_def = Brace(x, sharpness=0.7, direction=[-1,0,0]);
        self.play(Write(curly_brace_around_rec_def));
        self.play(Write(x), run_time=3)
        #self.wait(0.3)
        #self.play(Restore(self.camera.frame))

        self.add_example_of_basisfunctions()
        self.add_example_of_complete_basis()
        self.plot_data_and_bsplines()
        return
    
    
    def add_example_of_basisfunctions(self):
        ### ADD Axis and B-splines (constant, linear and cubic) example
        
        axes = Axes(x_range=[0,1.2,1], y_range=[0,1.2,1], x_length=5, y_length=2, 
                   axis_config={"include_tip": True, "numbers_to_exclude": [0]}).add_coordinates()
        
        axes.move_to([-14, 3, 0], aligned_edge=LEFT)
        
        self.play(Write(Text("Examples of B-spline basisfunctions", font_size=35).next_to(axes, UP, aligned_edge=LEFT, buff=0.5)))
        graph3 = axes.get_graph(lambda xx: BS.basisfunction(xx, knots, 9, 3), 
                               x_range=(0, 1, 0.005), color=RED, use_smoothing=False)
        graph1 = axes.get_graph(lambda xx: BS.basisfunction(xx, knots, 6, 1), 
                               x_range=(0, 1, 0.005), color=GREEN, use_smoothing=False)
        graph0 = axes.get_graph(lambda xx: BS.basisfunction(xx, knots, 4, 0), 
                               x_range=(0, 1, 0.005), color=YELLOW, use_smoothing=False)
        
        graphing_stuff = VGroup(graph0, graph1, graph3)
            
        self.play(DrawBorderThenFill(axes))
        self.add(Text("Constant", font_size=20, color=YELLOW).move_to(graph0.get_center()+1.2*LEFT+1.2*UP))
        self.play(Create(graph0, run_time=1.0))
        self.add(Text("Linear", font_size=20, color=GREEN).move_to(graph1.get_center()+1.1*UP))
        self.play(Create(graph1, run_time=1.0))
        self.add(Text("Cubic", font_size=20, color=RED).move_to(graph3.get_center()+RIGHT+0.8*UP))
        self.play(Create(graph3, run_time=1.0))
        
        return 
    
    def add_example_of_complete_basis(self):
        ### ADD Axis and B-splines linear cubic example
        
        axes = Axes(x_range=[0,1.2,1], y_range=[0,1.2,1], x_length=5, y_length=2, 
                   axis_config={"include_tip": True, "numbers_to_exclude": [0]}).add_coordinates()
        
        axes.move_to([-14, 0, 0], aligned_edge=LEFT)
        
        self.play(Write(Text("B-spline basis", font_size=35).next_to(axes, UP, aligned_edge=LEFT, buff=0.25)))
        self.play(DrawBorderThenFill(axes))
        
        # Create knots for l=0
        NR_SPLINES = 10
        B, knots_0 = Bspline.basismatrix(np.linspace(0,1,1000), l=0).values()
        graphs_0 = []
        for i in range(NR_SPLINES):
            graphs_0.append(
                axes.get_graph(lambda xx: BS.basisfunction(xx, knots_0, i, 0), 
                               x_range=(0, 1, 0.005), color=GRAY, use_smoothing=False, stroke_width=2)
            )
        B, knots_1 = Bspline.basismatrix(np.linspace(0,1,1000), l=1).values()
        graphs_1 = []
        for i in range(NR_SPLINES):
            graphs_1.append(
                axes.get_graph(lambda xx: BS.basisfunction(xx, knots_1, i+1, 1), 
                               x_range=(0, 1, 0.005), color=GRAY, use_smoothing=False, stroke_width=2)
            )
        
        B, knots_3 = Bspline.basismatrix(np.linspace(0,1,1000), l=3).values()
        graphs_3 = []
        for i in range(NR_SPLINES):
            graphs_3.append(
                axes.get_graph(lambda xx: BS.basisfunction(xx, knots_3, i+3, 3), 
                               x_range=(0, 1, 0.005), color=GRAY, use_smoothing=False, stroke_width=2)
            )
            
        CONSTANT_BASIS = Text("Constant, l=0", font_size=20, color=GRAY).move_to(axes.get_center()+1.1*UP) 
        LINEAR_BASIS = Text("Linear, l=1", font_size=20, color=GRAY).move_to(axes.get_center()+1.1*UP)
        CUBIC_BASIS = Text("Cubic, l=3", font_size=20, color=GRAY).move_to(axes.get_center()+1.1*UP)
        
        self.add_and_remove_basis(graphs_0, CONSTANT_BASIS)
        self.add_and_remove_basis(graphs_1, LINEAR_BASIS)
        self.add_and_remove_basis(graphs_3, CUBIC_BASIS, REMOVE=False)
        
        return 
                        
                        
    def add_and_remove_basis(self, GRAPH_LIST, TITLE, WAIT_TIME=1, REMOVE=True):
        """ adds TITLE, plots the basis given in GRAPH_LIST, then removes TITLE """ 
        self.play(Write(TITLE))
        for g in GRAPH_LIST:
            self.play(Create(g, run_time=0.4))
        self.wait(WAIT_TIME)
        if REMOVE:
            for g in GRAPH_LIST:
                self.play(Uncreate(g, run_time=0.1))
            self.remove(TITLE)
        self.wait(WAIT_TIME)
        return
              
    def plot_data_and_bsplines(self):
        ## CREATE DATA AND PLOT IT ######################################################
        self.play(Create(MathTex(r"f(x) = \sqrt{1 + \sin(2\pi x)}^{5/2}", font_size=35).move_to([-3,4,0], aligned_edge=LEFT+UP)))
        axes = Axes(x_range=[0,1.2,0.2], y_range=[-0.5,6,1], x_length=7, y_length=5, 
                   axis_config={"include_tip": True, "numbers_to_exclude": [0]}).add_coordinates()
        axes.move_to([-7,4,0], aligned_edge=LEFT+UP)
        self.play(DrawBorderThenFill(axes))

        x = np.linspace(0,1,35)
        y = (1 + np.sin(2*np.pi*x))**2.5 + np.random.normal(loc=0.0, scale=0.2, size=len(x))

        data_points = axes.get_line_graph(x_values = x, y_values = y, line_color=BLACK, 
                                         vertex_dot_style=dict(stroke_width=1,  fill_color=PURPLE), stroke_width = 2)
        self.play(Create(data_points, run_time=3))
        self.wait(2)
        ## CREATE BSPLINE ESTIMATE AND PLOT IT #########################################

        BS = Bspline()
        # constant 
        coef_, basis, knots = Bspline.fit(x, y, nr_splines=10, l=0, knot_type="e").values()
        constBspline_fit = axes.get_line_graph(x_values = x, y_values = basis @ coef_, line_color=YELLOW, vertex_dot_radius=0, stroke_width=5)
        leg_const = Text("Constant", color=self.color_const, font_size=20).move_to([-2,2,0])
        self.play(Write(leg_const))
        self.play(Write(constBspline_fit, run_time=3))
        # linear
        self.wait(1)
        coef_, basis, knots = Bspline.fit(x, y, nr_splines=10, l=1, knot_type="e").values()
        linBspline_fit = axes.get_line_graph(x_values = x, y_values = basis @ coef_, line_color=GREEN, vertex_dot_radius=0, stroke_width=5)
        leg_lin = Text("Linear", color=self.color_linear, font_size=20).move_to([-2,1.7,0])
        self.play(Write(leg_lin))
        self.play(Write(linBspline_fit, run_time=3))
        # cubcic
        self.wait(1)
        coef_, basis, knots = Bspline.fit(x, y, nr_splines=10, l=3, knot_type="e").values()
        cubBspline_fit = axes.get_line_graph(x_values = x, y_values = basis @ coef_, line_color=RED, vertex_dot_radius=0, stroke_width=5)
        leg_cub = Text("Cubic", color=self.color_cubic, font_size=20).move_to([-2,1.4,0])
        self.play(Write(leg_cub))
        self.play(Write(cubBspline_fit, run_time=3))
        return
    
    

        """
        
        
        # move to square
        self.play(self.camera.frame.animate.set(width=square.width * 2).move_to(square))
        self.remove(square)
        self.play(self.camera.frame.animate.set(width=square.width * 2).move_to(square))
        self.play(Create(Text("B-splines", font_size=40, color=BLUE).move_to([-6,4,0])))
        self.wait(0.3)
        self.play(Restore(self.camera.frame))
        
        # move to A_spline
        self.play(self.camera.frame.animate.set(width=circle.width * 2).move_to(A_spline))
        self.remove(A_spline)
        self.play(self.camera.frame.animate.set(width=A_spline.width * 2).move_to(A_spline))
        self.play(Create(Text("A-splines", font_size=40, color=BLUE).move_to([-6,-4,0])))
        self.wait(0.3)
        self.play(Restore(self.camera.frame))
        
        # move to triangle
        self.play(self.camera.frame.animate.set(width=circle.width * 2).move_to(triangle))
        self.remove(triangle)
        self.play(self.camera.frame.animate.set(width=triangle.width * 2).move_to(triangle))
        self.play(Create(Text("TP-B-splines", font_size=40, color=BLUE).move_to([6,-4,0])))

        self.wait(0.3)
        
        """
```

```python
%%manim -v WARNING -qm Grid
class Grid(Scene):
    """
    Manim uses a grid to place elements. At the center of the image is the origin, coordinate (0, 0, 0).
    This example demonstrates the size of the grid.
    """

    def construct(self):
        dots = []
        for x in range(-7, 8):
            for y in range(-5, 6):
                dot = Dot(point=(x, y, 0))
                dots.append(dot)
        self.add(*dots)
        self.add(Dot(color=RED))
```

```python
%%manim -v WARNING -qm GraphingIntro
class GraphingIntro(Scene):
    def construct(self):

        backg_plane = NumberPlane(x_range=[-7,7,1], y_range=[-4,4,1]).add_coordinates()
        code = Code("Tute2Intro.py", style=Code.styles_list[12], background ="window", language = "python", insert_line_no = True,
        tab_width = 2, line_spacing = 0.3, font_size = 0.5, font="Monospace").set_width(7).to_edge(UL, buff=0)

        axes = Axes(x_range = [0,5,1], y_range = [0,3,1], 
        x_length = 5, y_length = 3,
        axis_config = {"include_tip": True, "numbers_to_exclude": [0]}
        ).add_coordinates()

        axes.to_edge(UR)
        axis_labels = axes.get_axis_labels(x_label = "x", y_label = "f(x)")

        graph = axes.get_graph(lambda x : x**0.5, x_range = [0,4], color = YELLOW)

        graphing_stuff = VGroup(axes, graph, axis_labels)

        self.play(FadeIn(backg_plane), Write(code), run_time=6)
        self.play(backg_plane.animate.set_opacity(0.3))
        self.wait()
        self.play(DrawBorderThenFill(axes), Write(axis_labels), run_time = 2)
        self.wait()
        self.play(Create(graph), run_time = 2)
        self.play(graphing_stuff.animate.shift(DOWN*4), run_time = 3)
        self.wait()
        self.play(axes.animate.shift(LEFT*3), run_time = 3)
        self.wait()
```

```python
%%manim -v WARNING -ql Tute1
class Tute1(Scene): #ILLUSTRATING HOW TO PUT A NUMBER PLANE ON SCENE WITH A GRAPH, and a line using c2p
    def construct(self):

        backg_plane = NumberPlane(x_range=[-7,7,1], y_range=[-4,4,1])
        backg_plane.add_coordinates()
        #code = Code("Tute2Code2.py", style=Code.styles_list[12], background ="window", language = "python", insert_line_no = True,
        #tab_width = 2, line_spacing = 0.3, scale_factor = 0.5, font="Monospace").set_width(7).to_edge(UL, buff=0)

        my_plane = NumberPlane(x_range = [-6,6], x_length = 5,
        y_range = [-10,10], y_length=5)
        my_plane.add_coordinates()
        my_plane.shift(RIGHT*3)

        my_function = my_plane.get_graph(lambda x : 0.1*(x-5)*x*(x+5), 
        x_range=[-6,6], color = GREEN_B)

        label = MathTex("f(x)=0.1x(x-5)(x+5)").next_to(
            my_plane, UP, buff=0.2)

        area = my_plane.get_area(graph = my_function, 
        x_range = [-5,5], color = [BLUE,YELLOW])

        horiz_line = Line(
            start = my_plane.c2p(0, my_function.underlying_function(-2)),
        end = my_plane.c2p(-2, my_function.underlying_function(-2)),
        stroke_color = YELLOW, stroke_width = 10)

        #self.play(FadeIn(backg_plane), Write(code), run_time=6)
        self.play(FadeIn(backg_plane), run_time=6)
        self.play(backg_plane.animate.set_opacity(0.2))
        self.wait()
        self.play(DrawBorderThenFill(my_plane), run_time=2)
        self.wait()
        self.play(Create(my_function), Write(label), run_time=2)
        self.wait()
        self.play(FadeIn(area), run_time = 2)
        self.wait()
        self.play(Create(horiz_line), run_time = 2)
        self.wait()
```

```python
class Tute3(Scene): #Showing how to call 2 planes, put graphs on each and call elements to each
    def construct(self):

        backg_plane = NumberPlane(x_range=[-7,7,1], y_range=[-4,4,1]).add_coordinates()

        plane = NumberPlane(x_range = [-4,4,1], x_length = 4, 
        y_range= [0, 20, 5], y_length=4).add_coordinates()
        plane.shift(LEFT*3+DOWN*1.5)
        plane_graph = plane.get_graph(lambda x : x**2, 
        x_range = [-4,4], color = GREEN)
        area = plane.get_riemann_rectangles(graph = plane_graph, x_range=[-2,2], dx=0.05)

        axes = Axes(x_range = [-4,4,1], x_length = 4, 
        y_range= [-20,20,5], y_length=4).add_coordinates()
        axes.shift(RIGHT*3+DOWN*1.5)
        axes_graph = axes.get_graph(lambda x : 2*x, 
        x_range=[-4,4], color = YELLOW)
        v_lines = axes.get_vertical_lines_to_graph(
            graph = axes_graph, x_range=[-3,3], num_lines = 12)

        code = Code("Tute2Code1.py", style=Code.styles_list[12], background ="window", language = "python", insert_line_no = True,
        tab_width = 2, line_spacing = 0.3, scale_factor = 0.5, font="Monospace").set_width(6).to_edge(UL, buff=0)

        self.play(FadeIn(backg_plane), Write(code), run_time=6)
        self.play(backg_plane.animate.set_opacity(0.3))
        self.play(Write(plane), Create(axes))
        self.wait()
        self.play(Create(plane_graph), Create(axes_graph), run_time = 2)
        self.add(area, v_lines)
        self.wait()

class Tute4(Scene):
    def construct(self):

        plane = ComplexPlane(axis_config = {"include_tip": True, "numbers_to_exclude": [0]}).add_coordinates()

        labels = plane.get_axis_labels(x_label = "Real", y_label="Imaginary")

        quest = MathTex("Plot \\quad 2-3i").add_background_rectangle().to_edge(UL)
        dot = Dot()
        vect1 = plane.get_vector((2,0), stroke_color = YELLOW)
        vect2 = Line(start = plane.c2p(2,0),
        end = plane.c2p(2,-3), stroke_color = YELLOW).add_tip()

        self.play(DrawBorderThenFill(plane), Write(labels))
        self.wait()
        self.play(FadeIn(quest))
        self.play(GrowA_spline(vect1), dot.animate.move_to(plane.c2p(2,0)), rate_func = linear, run_time = 2)
        self.wait()
        self.play(GrowFromPoint((vect2), point = vect2.get_start()),
        dot.animate.move_to(plane.c2p(2,-3)), run_time = 2, rate_func = linear)
        self.wait()
```
