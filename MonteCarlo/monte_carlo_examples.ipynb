## Coin Flip Example

$ P(Head) = P(Tail) = 1/2$

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def coin_flip():
    return random.randint(0,1)
coin_flip(), coin_flip(), coin_flip()
```

```python
def monte_carlo_coin_flip(n):
    list1 = []

    results = 0
    for i in range(n):
        flip_result = coin_flip()
        results += flip_result
        
        # Calculate probabiltiy value
        prob_value = results / (i+1)
        # Append the probability value to the list
        list1.append(prob_value)
        # Plot the results
        
    return results/n, list1  
```

```python
answer, list1 = monte_carlo_coin_flip(n=500000)
print("Final value:", answer)

fig = plt.figure()
plt.axhline(y=0.5, color='r', linestyle='-')
plt.xlabel("Iteration")
plt.ylabel('Probability')
plt.ylim(0.4975, 0.5025)
plt.grid()
plt.plot(list1)
```

## Estimating $\pi$ using circle and square

To estimate the value of $\pi$, we need the area of a square and of a circle. Place random dots on the sruface and count the dots that fall inside the circle and the dots that fall inside the square. Such will give an estimated amount of the areas. Therefore, instead of using the actual areas, we will use the count of dots to use as areas. 

$\frac{\text{Area of square}}{\text{Area of Circle}} = \frac{(2r)^2}{\pi r^2} = \frac{4}{\pi}$

![Value of pi using Monte Carlo Simulation](MC_valueOfPi.png)

```python
import turtle 
import random
import matplotlib.pyplot as plt
import math
```

```python
myPen = turtle.Turtle()
```

```python
myPen.hideturtle()
myPen.speed(0)

# Drawing a square
myPen.up()
myPen.setposition(-100, -100)
myPen.down()
myPen.fd(200)
myPen.left(90)
myPen.fd(200)

myPen.left(90)
myPen.fd(200)
myPen.left(90)
myPen.fd(200)
myPen.left(90)

# Drawing a circle
myPen.up()
myPen.setposition(0, -100)
myPen.down()
myPen.circle(100)
```

```python
# To count the points inside/outside the circle
in_circle = 0
out_circle = 0

# To store the values of Pi
pi_values = []

PLOT = False
for j in range(5000):
    # generate random numbers
    x = random.randrange(-100, 100)
    y = random.randrange(-100, 100)
    
    # Check if the number lies outside the circel
    if (x**2 + y**2 > 100**2):
        if PLOT:
            myPen.color("black")
            myPen.up()
            myPen.goto(x,y)
            myPen.down()
            myPen.dot()
        out_circle += 1
    else:
        if PLOT:
            myPen.color("red")
            myPen.up()
            myPen.goto(x,y)
            myPen.down()
            myPen.dot()
        in_circle += 1
    
    # Calculate the value of pi
    pi = 4 * in_circle / (in_circle + out_circle)
    
    pi_values.append(pi)
    
    # Calculate the errors
    avg_pi_erros = [abs(math.pi - pi) for pi in pi_values]
    
print(pi_values[-1])
```

```python
# Plot the pi values
fig = plt.figure()
plt.axhline(y=math.pi, color="g", linestyle="-")
plt.xlabel("Iteration")
plt.ylabel("Value of pi")
plt.plot(pi_values)
fig.show()
fig = plt.figure()
plt.axhline(y=0.0, color="g", linestyle="-")
plt.plot(avg_pi_erros)
plt.xlabel("Iteration")
plt.ylabel("Error")
fig.show()
```

## Monty Hall Problem
Suppose you are on a game show, and you have the choice of picking one of three doors: Behind one door is a car; behind the other doors, goats.  
You pick a door, let’s say door 1, and the host, who knows what’s behind the doors, opens another door, say door 3, which has a goat. The host then asks you: do you want to stick with your choice or choose another door?  
Is it to your advantage to switch your choice of door?



![Monty Hall](monty_hall.png)

```python
import random
import matplotlib.pyplot as plt
```

```python
doors = ["goat", "goat", "car"]

# empty list to store probability values
switch_win_probability = []
stick_win_probability = []
```

```python
def monte_carlo_montyhall(n):
    
    # calculate switch and stick wins
    switch_wins = 0
    stick_wins = 0
    for i in range(n):
        # randomly placeing the car and goats behind the tree doors
        random.shuffle(doors)
        # contestant's choice
        k = random.randrange(2)
        # If contestent doesnt get the car
        if doors[k] != "car":
            switch_wins += 1
        # If contestant got car
        else:
            stick_wins += 1
        
        # Update the list values
        switch_win_probability.append(switch_wins/(i+1))
        stick_win_probability.append(stick_wins/(i+1))
```

```python
monte_carlo_montyhall(1000)
fig = plt.figure()
plt.axhline(y = 2/3, color="r", linestyle="-")
plt.axhline(y = 1/3, color="g", linestyle="-")
plt.plot(switch_win_probability, "b", label="Switch wins")
plt.plot(stick_win_probability, "k", label="Stick wins")
```

## Buffon's Needle Problem

A French nobleman Georges-Louis Leclerc, Comte de Buffon, posted the following problem in 1777.

'''
*Suppose that we drop a short needle on a ruled paper — what would be the probability that the needle comes to lie in a position where it crosses one of the lines?*
'''

The probability depends on the distance (d) between the lines of the ruled paper, and it depends on the length (l) of the needle that we drop — or rather, it depends on the ratio l/d. For this example, we can interpret the needle as l ≤ d. In short, our purpose is that the needle cannot cross two different lines at the same time. Surprisingly, the answer to the Buffon’s needle problem involves PI.

Here we are going to use the solution of Buffon’s needle problem to estimate the value of PI experimentally using the Monte Carlo Method. However, before going into that, we are going to show how the solution derives, making it more interesting.

#### Theorem:
If a short needle, of length l, is dropped on a paper that is ruled with equally spaced lines of distance d ≥ l, then the probability that the needle comes to lie in a position where it crosses one of the lines is:

$$ P = \frac{2l}{\pi d}$$

![Buffon's Needle Problem](buffons_needle.png)

```python
import random
import math
import matplotlib.pyplot as plt
```

```python
def monte_carlo_buffon(runs, needles, n_length, b_width):
    pi_values = []
    
    for i in range(runs):
        nhits = 0
        for j in range(needles):
            x = random.uniform(0, b_width/2.0)
            theta = random.uniform(0, math.pi/2)
            
            xtip = x - (n_length/2.0)*math.cos(theta)
            if xtip < 0:
                nhits += 1
        numerator = 2.0 * n_length * needles
        denominator = b_width * nhits
        pi_values.append(numerator / denominator)
    print(pi_values[-1])
    
    fig = plt.figure()
    plt.axhline(y=math.pi, color="r", linestyle="-")
    plt.plot(pi_values)
```

```python
runs = 100
needles = 100000
n_length = 2
b_width = 2
monte_carlo_buffon(runs, needles, n_length, b_width)
```
