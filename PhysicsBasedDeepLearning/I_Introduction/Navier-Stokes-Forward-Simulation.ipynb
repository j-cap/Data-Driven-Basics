## Forward simulation of the Navier-Stokes Equation with $\phi$Flow

$$\frac{\partial u}{\partial t} + u * \nabla u = -\frac{1}{\rho} \nabla \rho + \nu \nabla * \nabla u + (0,1)^T \xi d \quad s.t. \nabla * u = 0$$
$$\frac{\partial d}{\partial t} + u * \nabla d = 0$$  
with Dirichlet boundary conditions $u = 0$ for the velocity and Neuman boundaries $\frac{\partial}{\partial x} = 0$ for pressure, on a domain $\Omega$ with a physical sicze of $100 \times 80$ units

```python
import numpy as np
import pylab
from phi import flow
```

```python
DT = 1.5
NU = 0.01

INFLOW = flow.CenteredGrid(flow.Sphere(center=(30,15), radius=10), flow.extrapolation.BOUNDARY, x=32, y=40, bounds=flow.Box[0:80, 0:100])*0.2
```

```python
smoke = flow.CenteredGrid(0, flow.extrapolation.BOUNDARY, x=32, y=40, bounds=flow.Box[0:80, 0:100])
velocity = flow.StaggeredGrid(0, flow.extrapolation.ZERO, x=32, y=40, bounds=flow.Box[0:80, 0:100])
```

```python
# define update step
def step(velocity, smoke, pressure, dt=1.0, buoyancy_factor=1.0):
    smoke = flow.advect.semi_lagrangian(smoke, velocity, dt) + INFLOW
    buoyancy_force = smoke * (0, buoyancy_factor) >> velocity # resample smoke to velocity sample points
    velocity = flow.advect.semi_lagrangian(velocity, velocity, dt) + dt*buoyancy_force
    velocity = flow.diffuse.explicit(velocity, NU, dt)
    velocity, pressure = flow.fluid.make_incompressible(velocity)
    return velocity, smoke, pressure
```

```python
velocity, smoke, pressure = step(velocity, smoke, None, dt=DT)

print('Max. velocity and mean marker density: '+format( [flow.math.max(velocity.values), flow.math.mean(smoke.values)]))
```

```python
# time evolution
for time_step in range(10):
    velocity, smoke, pressure = step(velocity, smoke, pressure, dt=DT)
    print('Computed frame {} , max velocity {} '.format(time_step , np.asarray(flow.math.max(velocity.values)) ))
```

```python
pylab.imshow(smoke.values.numpy('y,x'), origin='lower', cmap='magma')
```

```python
# more time evolution
steps = [[ smoke.values, velocity.values.vector[0], velocity.values.vector[1] ]]
for time_step in range(50):
    if time_step < 3 or time_step%10 == 0:
        print('Computing time step %d' % time_step)
    velocity, smoke, pressure = step(velocity, smoke, pressure, dt=DT)
    if time_step%5 == 0:
        steps.append( [smoke.values, velocity.values.vector[0], velocity.values.vector[1]] ) 
        
```

```python
fig, axes = pylab.subplots(1, len(steps), figsize=(16, 5))
for i in range(len(steps)):
    axes[i].imshow(steps[i][0].numpy('y,x'), origin='lower', cmap='magma')
    axes[i].set_title(f"d at t={ i*5} ")
```
