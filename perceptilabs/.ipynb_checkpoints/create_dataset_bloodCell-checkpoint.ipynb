
Creates a labeled dataset from the Bloodcell dataset in 
     C:\Users\WeberJ\Documents\Data\dataset2-master\dataset2-master


Possible labels:
- Eosinophil
- Lymphocyte
- Monocyte
- Neutrophil

```python
from pathlib import Path
import os

EOSINOPHIL = 0
LYMPHOCYTE = 1
MONOCYTE = 2
NEUTROPHIL = 3

TYPES = {"EOSINOPHIL":0, "LYMPHOCYTE":1, "MONOCYTE":2, "NEUTROPHIL":3}

DATA_PATH = Path('C:\\Users\\WeberJ\\Documents\\Data\\dataset2-master\\dataset2-master')
print(TYPES)
```

```python
# create data.csv with image_paths and labels
with open("data.csv", "w") as f:
    headers = "id,label"
    f.write(headers)
    f.write("\n")

    for folder in os.listdir(DATA_PATH / 'images/'):
        for type_ in TYPES:
            print(type_, " label = ", TYPES[type_])

            for img in os.listdir(DATA_PATH / 'images/' / folder / type_):
                label = TYPES[type_]
                id_ = img
                line = id_ + "," + str(label)
                f.write(line)
                f.write("\n")
```

```python
for folder in os.listdir(DATA_PATH / 'images/'):
    for type_ in TYPES:
        print(type_, " label = ", TYPES[type_])

        for img in os.listdir(DATA_PATH / 'images/' / folder / type_):
            print(img)
```

```python
from pathlib import Path
import shutil

src_path = DATA_PATH / "images"
trg_path = DATA_PATH / "images"
```

```python
for src_file in Path(src_path).glob('*.*'):
    shutil.copy(src_file, trg_path)
```
